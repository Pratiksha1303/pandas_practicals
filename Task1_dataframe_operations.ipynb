=> Create two dictionaries, each containing the keys 'ID', 'Name', and 'Class', with four records under each key.

=>Convert both dictionaries into two separate DataFrames, and then perform concatenation and join operations on these DataFrames.  

```python
import pandas as pd
```

```python
# Dictionary 1

data1 = {
    "ID": [1, 2, 3, 4],
    "Name": ["Ana", "Bob", "Charlie", "Julie"],
    "Class": ["CSE", "AI", "DS", "CSE"]
}

# Dictionary 2

data2 = {
    "ID": [5, 6, 7, 8],
    "Name": ["Elena", "Harry", "John", "Alison"],
    "Class": ["Cyber", "AI", "DS", "CSE"]
}
```

```python
# convert to DataFrames

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
print("DataFrame 1")
print(df1)
```

```python
print("DataFrame 2")
print(df2)
```

```python
# Concatention

concat_df = pd.concat([df1, df2])

print("\nConcatenated DataFrame")
print("\n",concat_df)
```

```python
# Join Opeartion

joined_df = df1.join(df2, lsuffix='_left', rsuffix='_right')

print("\nJoined DataFrame")
print(joined_df)
```
