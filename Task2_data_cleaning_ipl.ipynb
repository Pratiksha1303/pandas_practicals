=> To clean data and handle missing values in the IPL dataset.  

```python
import pandas as pd
```

```python
# Load Dataset

df = pd.read_csv("IPL_Matches_2008_2022.csv")
```

```python
# show first 5 rows

print(df.head(5))
```

```python
# check  missing values

print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Fill missing values with mean(for numeric columns)

df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Fill missing values in object columns

df.fillna("Unknow", inplace=True)
```

```python
# Remove duplicate rows

df.drop_duplicates(inplace=True)
```

```python
#Check again

print("\nAfter Cleaning:")
print(df.isnull().sum())
```
